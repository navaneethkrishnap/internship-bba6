# Data Visualization

## Visitors over time

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.figure(figsize=(12,8))
x,y = df['Date'],df['Visitors']
plt.plot(x,y,marker='o',linestyle='-',color='b',label='Monthly Traffic')
plt.xlabel('Date')
plt.ylabel('Monthly Visitors')
plt.title('Monthly Visitors for Year 2023-2024')
plt.xticks(rotation=None)
plt.legend()
# plt.grid()
plt.show()
```

## Unique Visitors vs Visitors ( Stacked Bar Chart )

```python
df.set_index('Date')[['Visitors', 'Unique Visitors']].plot(kind='bar', stacked=True)
plt.title('Cumulative Website Data')
plt.ylabel('Count')
plt.show()
```

## Visitors vs Bounce Rate Correlation

```python
plt.scatter(df['Visitors'], df['Bounce Rate (%)'])
plt.xlabel('Visitors')
plt.ylabel('Bounce Rate (%)')
plt.title('Visitors vs Bounce Rate')
plt.show()

# No correlation between Visitors and Bounce Rate but can see pretty constant
```

## Correlation Matrix shown using Heatmaps

```python
import seaborn as sns
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

##  Types of Refferral ( % ) - Pie Chart

```python
p = df['Paid Traffic (%)'].sum()
o = df['Organic Traffic (%)'].sum()
r = df['Referral Traffic (%)'].sum()
d = df['Direct Traffic (%)'].sum()

# Data for pie chart
labels = ['Paid Traffic', 'Organic Traffic', 'Referral Traffic', 'Direct Traffic']
sizes = [p, o, r, d]
colors = ['green', 'blue', 'red', 'purple']

# Plot the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Proportion of Traffic Sources (Total)')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Summing up traffic sources
p = df["Paid Traffic (%)"].sum()
o = df["Organic Traffic (%)"].sum()
r = df["Referral Traffic (%)"].sum()
d = df["Direct Traffic (%)"].sum()

# Data for pie chart
labels = ["Paid Traffic", "Organic Traffic", "Referral Traffic", "Direct Traffic"]
sizes = [p, o, r, d]

# Use the same pastel color palette
colors = sns.color_palette("pastel")

# Create a full pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    sizes,
    labels=labels,
    autopct="%1.1f%%",
    colors=colors,
    startangle=140,
    wedgeprops={"edgecolor": "white", "linewidth": 1, "antialiased": True},  # Smooth edges
    textprops={"fontsize": 12, "color": "black"},  # No boldness, clean text
)

# Title
plt.title("Proportion of Traffic Sources", fontsize=14)

# Ensure it's a perfect circle
plt.axis("equal")

# Show the plot
plt.show()
```
